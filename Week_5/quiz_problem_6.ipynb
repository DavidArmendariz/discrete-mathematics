```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edge('A','B')
G.add_edge('A','C')
G.add_edge('A','D')
G.add_edge('A','E')
G.add_edge('B','C')
G.add_edge('B','D')
G.add_edge('B','E')
G.add_edge('C','D')
G.add_edge('C','E')
G.add_edge('D','E')
G.add_node('F')
fig = plt.figure(figsize=(5,5))
nx.draw_networkx(G)
plt.savefig('disconnected.png')
plt.show()
```
