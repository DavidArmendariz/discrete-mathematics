```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
mygraph = nx.Graph()
mygraph.add_edge('A', 'B')
mygraph.add_edge('B', 'C')
mygraph.add_edge('C', 'D')
mygraph.add_edge('D', 'E')
mygraph.add_edge('E', 'F')
mygraph.add_edge('F', 'G')
mygraph.add_edge('G', 'A')
```

```python
nx.draw_networkx(mygraph)
plt.savefig("first_case.png", format="PNG")
plt.show()
```

```python
mygraph = nx.Graph()
mygraph.add_edge('A', 'B')
mygraph.add_edge('B', 'C')
mygraph.add_edge('C', 'A')
mygraph.add_edge('D', 'E')
mygraph.add_edge('E', 'F')
mygraph.add_edge('F', 'G')
mygraph.add_edge('G', 'D')
```

```python
nx.draw_networkx(mygraph)
plt.savefig("second_case.png", format="PNG")
plt.show()
```
